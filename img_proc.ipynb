```python
first_channel = avatar[:,:,0]
second_channel = avatar[:,:,1]
third_channel = avatar[:,:,2]

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(13, 3), ncols=3)
ax1.imshow(first_channel)
ax2.imshow(second_channel)
ax3.imshow(third_channel)
```

### *Let`s test color conversion to hex on one pixel.*

```python
webcolors.rgb_to_hex((round(first_channel[0,0]), round(second_channel[0,0]), round(third_channel[0,0])))
```

```python
hex_array = []
for i in range(avatar.shape[0]):
    for j in range(avatar.shape[1]):
        hex_pixel_code = webcolors.rgb_to_hex((round(first_channel[i,j]), round(second_channel[i,j]), round(third_channel[i,j])))
        hex_array.append(hex_pixel_code)
```

Let`s find the median of our image array

```python
median1 = np.median(first_channel, axis=1)
median2 = np.median(second_channel, axis=1)
median3 = np.median(third_channel, axis=1)
```

```python
hex_array2 = []
for i in range(median1.shape[0]):
    hex_pixel_code = webcolors.rgb_to_hex((round(median1[i]), round(median2[i]), round(median3[i])))
    hex_array2.append(hex_pixel_code)
```

```python
sorted_hex_array2 = sorted(hex_array2)
```
