```python
avatar = io.imread("00000.jpg")
print(avatar.shape) # geometry
print(avatar.size) # the number of pixels
plt.imshow(avatar)
```

```python
first_channel = avatar[:,:,0]
second_channel = avatar[:,:,1]
third_channel = avatar[:,:,2]

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(13, 3), ncols=3)
ax1.imshow(first_channel)
ax2.imshow(second_channel)
ax3.imshow(third_channel)
```

### *Let`s test color conversion to hex on one pixel.*

```python
webcolors.rgb_to_hex((first_channel[0,0], second_channel[0,0], third_channel[0,0]))
```

```python
hex_array = []
for i in range(avatar.shape[0]):
    for j in range(avatar.shape[1]):
        hex_pixel_code = webcolors.rgb_to_hex((first_channel[i,j], second_channel[i,j], third_channel[i,j]))
        hex_array.append(hex_pixel_code)
```

```python
len(hex_array), avatar.size/3
```

```python
rgb_array = []
for i in range(avatar.shape[0]):
    for j in range(avatar.shape[1]):
        rgb_array.append(sum((first_channel[i,j], second_channel[i,j], third_channel[i,j])))
```

```python
unique_rgb = np.array(list(set(rgb_array)))
# len(unique_rgb)
# print(unique_rgb)
mean_unique_rgb = np.mean(unique_rgb)
```
